```python
import pandas as pd
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')  # Replace 'localhost' and '27017' with your MongoDB host and port
db = client['DAP']  # Connect to your database
collection = db['Diabetes_Destination']  # Connect to your collection

# Query MongoDB collection and load data into a Pandas DataFrame
cursor = collection.find()  # Retrieve all documents from the collection
df = pd.DataFrame(list(cursor))  # Convert cursor to list and then to DataFrame

# Close MongoDB connection
client.close()

# Display the DataFrame
print(df.head())  # Display the first few rows of the DataFrame
```

```python
#Lets perform some data transformation first, as there is no need for cleaning that much
#this dataset of diabetes has some columns which needs to be transformed for better data understanding and analysis
#We have values for the variable age from 1 to 13, where 1 to 8 are 18 to 24, 9 to 12 are 60 to 64 and 13 are 80 or 80 plus

# Replacing the 'Age' values with categories
df['Age'] = df['Age'].replace({1: 'Young', 2: 'Young', 3: 'Adult', 4: 'Adult',
                                5: 'Adult', 6: 'Middle-aged', 7: 'Middle-aged', 8: 'Middle-aged',
                                9: 'Middle-aged', 10: 'Senior', 11: 'Senior', 12: 'Senior',
                                13: 'Senior'})
```

```python
df.Age
```

```python
#Now we have another column for Education,lets do transformation with that variable to make it more understandable 
# here 1 = Never attended school or only kindergarten 
#2 = Grades 1 through 8 (Elementary) 
#3 = Grades 9 through 11 (Some high school) 
#4 = Grade 12 or GED (High school graduate) 
#5 = College 1 year to 3 years (Some college or technical school) 6 = College 4 years or more (College graduate)
#lets transform Education level
import pandas as pd

# Defining a mapping dictionary for education levels
education_mapping = {
    1: 'Pre-Kindergarten',
    2: 'Elementary School',
    3: 'Some High School',
    4: 'High School Graduate',
    5: 'Some College',
    6: 'College Graduate'
}

# Replacing education values with corresponding categories
df['Education'] = df['Education'].replace(education_mapping)
```

```python
#Now there is another variable for Income as well, lets try to transform that variable as well
#On the basis of information that this data is having, lets categories Low, Medium and High
# Replacing the 'Age' values with categories
df['Income'] = df['Income'].replace({1: 'Low', 2: 'Low', 3: 'Medium', 4: 'Medium',
                                5: 'Medium', 6: 'High', 7: 'High', 8: 'High'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
def count_values(dataset, cat_variable, order=None):
    """
    Function: Counts values in each category and displays them on a plot.

    Parameters: Dataset, category feature, and order of appearance (order is optional).
    """
    ax = sns.countplot(x=cat_variable, data=dataset, palette="husl", order=order)
    for p in ax.patches:
        ax.annotate(f"\n{p.get_height()}", (p.get_x()+0.2, p.get_height()),
                    ha="center", va="top", color="white", size=10)

    plt.title(f"Number of items in each {cat_variable} category")
    plt.show()

# Example usage:
# count_values(your_dataset, "category_variable", order=None)
```

```python
def plot_distribution(dataset, feature):
    """
    Function: Computes and displays distribution of features with continuous values; plots their mean and median.
    
    Parameters: Dataset and feature with continuous values.
    """
    plt.hist(dataset[feature], bins = "fd")
    
    plt.axvline(dataset[feature].mean(), color = "red", label = "mean")
    plt.axvline(dataset[feature].median(), color = "orange", label = "median")
    
    plt.xlabel(f"{feature}")
    plt.ylabel("Count")
    plt.legend()
    plt.title(f"Distribution of values in {feature}")
    plt.show()
```

```python
#Now lets load our data into its destination (Postgresql)
from sqlalchemy import create_engine

# Connect to PostgreSQL using SQLAlchemy
engine = create_engine('postgresql://postgres:Deadpool123#@localhost:5432/postgres')

# Replacing 'table_name' with our desired name to give us table in the PostgreSQL database
table_name = "DiabeteData"

try:
    # Write DataFrame to PostgreSQL using exception
    df.to_sql(table_name, con=engine, index=False, if_exists="replace")
    print("DataFrame has been successfully loaded into PostgreSQL.")
except Exception as e:
    print("An error occurred:", e)
finally:
    # Dispose the engine
    engine.dispose()
```
